# Transformas Columnas de Fechas a DateTime

Convertimos las columnas que contienen fechas a DateTime para poder trabajar mejor con ellas y realizar análisis de series temporales.

```python
# Crear una copia del DataFrame original
df_copy = df.copy()

# Lista de las columnas que contienen fechas
fecha_columns = ['FEC_PRIMERA_VENTA', 'FEC_ESTADO_DESDE', 'FEC_CREADO_CUENTA', 
                 'SEG_ALIADO_FEC_CREACION', 'ULTIMO_CONTACTO', 'ULTIMA_VENTA', 
                 'ULTIMO_SEG_COMERCIAL']

# Convertir las columnas de fechas a tipo datetime
for col in fecha_columns:
    df_copy[col] = pd.to_datetime(df_copy[col], errors='coerce')  # 'coerce' convierte errores a NaT (Not a Time)

# Convertir la columna MES_AÑO_PRIMERA_VENTA a tipo fecha
df_copy['MES_AÑO_PRIMERA_VENTA'] = pd.to_datetime(df_copy['MES_AÑO_PRIMERA_VENTA'].astype(str), format='%Y%m')

# Formatear la columna para que solo muestre el año y mes
df_copy['MES_AÑO_PRIMERA_VENTA'] = df_copy['MES_AÑO_PRIMERA_VENTA'].dt.to_period('M')

# Verificar la conversión
df_copy.info()
```

### 1. **Diagnóstico del proceso de liderazgo y manejo de KPIs de los Gerentes Comerciales:**
   - ¿Cuál es el promedio de días sin contacto para cada tipo de partner, y cómo puede esto afectar la estrategia de seguimiento?




```python
# Calculamos el promedio de días sin contacto para cada tipo de partner
promedio_dias_sin_contacto = df_copy.groupby('TIPO_PARTNER')['DIAS'].mean().sort_values(ascending=False)
print(promedio_dias_sin_contacto)

# Gráfica del promedio de días sin contacto por tipo de partner
plt.figure(figsize=(10,6))
sns.barplot(x=promedio_dias_sin_contacto.index, y=promedio_dias_sin_contacto.values, palette="viridis")
plt.title('Promedio de días sin contacto por tipo de partner')
plt.xlabel('Tipo de Partner')
plt.ylabel('Promedio de Días sin Contacto')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Comentario:
# Un mayor promedio de días sin contacto en un tipo de partner puede indicar menor frecuencia de interacción,
# lo que podría requerir ajustar la estrategia de seguimiento para evitar desatención y mejorar la relación comercial.
```

- ¿Qué porcentaje de partners tiene un contacto nulo o insuficiente en los últimos 60 y 90 días, y qué medidas se podrían implementar para mejorar la eficiencia comercial?

```python
# Calculamos el porcentaje de partners con contacto nulo o insuficiente en los últimos 60 y 90 días
total_partners = len(df_copy)

# Partners sin contacto en los últimos 60 días
sin_contacto_60 = df_copy['SIN_CONTACTO_60_DIAS'].sum()
porcentaje_sin_contacto_60 = (sin_contacto_60 / total_partners) * 100

# Partners sin contacto en los últimos 90 días
sin_contacto_90 = df_copy['SIN_CONTACTO_90_DIAS'].sum()
porcentaje_sin_contacto_90 = (sin_contacto_90 / total_partners) * 100

print(f"Porcentaje de partners sin contacto en los últimos 60 días: {porcentaje_sin_contacto_60:.2f}%")
print(f"Porcentaje de partners sin contacto en los últimos 90 días: {porcentaje_sin_contacto_90:.2f}%")


# Gráfica de porcentaje de partners sin contacto en los últimos 60 y 90 días
porcentajes = [porcentaje_sin_contacto_60, porcentaje_sin_contacto_90]
labels = ['Sin contacto 60 días', 'Sin contacto 90 días']

plt.figure(figsize=(6, 4))
sns.barplot(x=labels, y=porcentajes, palette='Blues')
plt.ylabel('Porcentaje (%)')
plt.title('Porcentaje de partners sin contacto en los últimos 60 y 90 días')
plt.ylim(0, 100)
for i, v in enumerate(porcentajes):
    plt.text(i, v + 1, f"{v:.2f}%", ha='center')
plt.tight_layout()
plt.show()
```

- ¿Cuál es la tasa de conversión de partners activos versus inactivos, y cómo varía esta tasa entre las diferentes categorías?

```python
# Calculamos la tasa de conversión de partners activos vs inactivos por categoría

# Consideramos 'Activo' en la columna 'ESTADO_DISTRIBUIDOR' como partners activos
activos_por_categoria = df_copy[df_copy['ESTADO_DISTRIBUIDOR'] == 'Activo'].groupby('TIPO_PARTNER')['ID_CUENTA'].count()
inactivos_por_categoria = df_copy[df_copy['ESTADO_DISTRIBUIDOR'] != 'Activo'].groupby('TIPO_PARTNER')['ID_CUENTA'].count()

# Unimos ambos conteos en un DataFrame
conversion_df = pd.DataFrame({
    'Activos': activos_por_categoria,
    'Inactivos': inactivos_por_categoria
}).fillna(0)

# Calculamos la tasa de conversión (activos / total)
conversion_df['Total'] = conversion_df['Activos'] + conversion_df['Inactivos']
conversion_df['Tasa_Conversion'] = (conversion_df['Activos'] / conversion_df['Total']) * 100

print(conversion_df[['Activos', 'Inactivos', 'Tasa_Conversion']].round(2))

# Visualización de la tasa de conversión por categoría
plt.figure(figsize=(10,6))
sns.barplot(x=conversion_df.index, y=conversion_df['Tasa_Conversion'], palette="YlGnBu")
plt.title('Tasa de conversión de partners activos por categoría')
plt.xlabel('Tipo de Partner')
plt.ylabel('Tasa de Conversión (%)')
plt.ylim(0, 100)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 4. **Indicadores relevantes y toma de decisiones:**
   - ¿Cuáles son los KPIs más relevantes para evaluar la efectividad de los asesores comerciales y su relación con el tipo de partner?

```python
# Seleccionamos algunos KPIs relevantes para evaluar la efectividad de los asesores comerciales:
# - Promedio de días sin contacto (DIAS)
# - Promedio de días desde la última venta (DIAS_ULTIMA_VENTA)
# - Total de seguimientos realizados (CANTIDAD_SEGUIMIENTOS)
# - Porcentaje de partners activos (Tasa de conversión)

# Calculamos los KPIs por tipo de partner
kpis_df = pd.DataFrame({
    'Promedio_Dias_Sin_Contacto': df_copy.groupby('TIPO_PARTNER')['DIAS'].mean(),
    'Promedio_Dias_Ultima_Venta': df_copy.groupby('TIPO_PARTNER')['DIAS_ULTIMA_VENTA'].mean(),
    'Total_Seguimientos': df_copy.groupby('TIPO_PARTNER')['CANTIDAD_SEGUIMIENTOS'].sum(),
    'Tasa_Conversion_Activos': conversion_df['Tasa_Conversion']
})

print("KPIs relevantes por tipo de partner:")
print(kpis_df.round(2))

# Visualización de los KPIs más relevantes
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
kpis_df['Promedio_Dias_Sin_Contacto'].sort_values().plot(kind='bar', ax=axes[0,0], color='skyblue', title='Promedio de Días Sin Contacto')
axes[0,0].set_ylabel('Días')

kpis_df['Promedio_Dias_Ultima_Venta'].sort_values().plot(kind='bar', ax=axes[0,1], color='lightgreen', title='Promedio de Días Desde Última Venta')
axes[0,1].set_ylabel('Días')

kpis_df['Total_Seguimientos'].sort_values().plot(kind='bar', ax=axes[1,0], color='salmon', title='Total de Seguimientos Realizados')
axes[1,0].set_ylabel('Seguimientos')

kpis_df['Tasa_Conversion_Activos'].sort_values().plot(kind='bar', ax=axes[1,1], color='gold', title='Tasa de Conversión de Activos (%)')
axes[1,1].set_ylabel('Porcentaje (%)')

plt.tight_layout()
plt.show()
```

### 5. **Recomendaciones de mejora:**
   - ¿Qué tipo de partner tiene mayor propensión a quedar sin contacto por más de 90 días, y cómo se podría mejorar su gestión?

```python
# Calculamos el porcentaje de partners sin contacto por más de 90 días por tipo de partner
sin_contacto_90_por_tipo = df_copy.groupby('TIPO_PARTNER')['SIN_CONTACTO_90_DIAS'].sum()
total_por_tipo = df_copy['TIPO_PARTNER'].value_counts()
porcentaje_sin_contacto_90_por_tipo = (sin_contacto_90_por_tipo / total_por_tipo * 100).sort_values(ascending=False)

print("Porcentaje de partners sin contacto >90 días por tipo de partner:")
print(porcentaje_sin_contacto_90_por_tipo.round(2))

# Visualización
plt.figure(figsize=(10,6))
sns.barplot(x=porcentaje_sin_contacto_90_por_tipo.index, y=porcentaje_sin_contacto_90_por_tipo.values, palette="Reds")
plt.title('Porcentaje de partners sin contacto >90 días por tipo de partner')
plt.xlabel('Tipo de Partner')
plt.ylabel('Porcentaje sin contacto >90 días (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Comentario:
# El tipo de partner con mayor porcentaje indica mayor riesgo de desatención.
# Para mejorar la gestión, se recomienda implementar alertas automáticas, segmentar campañas de reactivación y priorizar visitas o llamadas a estos grupos.
```
